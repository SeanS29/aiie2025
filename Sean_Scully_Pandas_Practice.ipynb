```python
print(titanicDF.isnull().values.any()) #To check if there any missing values
```

```python
print(titanicDF.isnull().sum().sum()) #To know how many null values
```

```python
titanicDF['Cabin'] = titanicDF['Cabin'].fillna(0)
```

```python
titanicDF['Age'] = titanicDF['Age'].fillna(titanicDF['Age'].mean())
```

```python
titanicDF.iloc[:1000, :100000]
```

```python
if 'Name' == np.nan:
  titanicDF.drop(axis =1, inplace=True)
```

```python
titanicDF['Gender'] = titanicDF['Gender'].replace({'male': 'm', 'female': 'f'})
```

```python
titanicDF.iloc[:1000, :100000]
```

```python
titanicDF['Died/Lived'] = titanicDF['Survived'].apply(lambda x: 'Lived' if x == 1 else 'Died')
titanicDF.head(10)
```

```python
# @title Age

from matplotlib import pyplot as plt
titanicDF['Age'].plot(kind='hist', bins=20, title='Age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# prompt: Find the top 3 people with the highest age

titanicDF.nlargest(3, 'Age')
```

```python
pivot_table = pd.pivot_table(titanicDF, values='Age', index='Gender', aggfunc=np.mean)
pivot_table
```
